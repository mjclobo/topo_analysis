# ECCO stuff

```python
ecco_dir = '/scratch/cimes/ml1994/global_topo_analysis/data/ecco_data/'

os.chdir(ecco_dir)
ecco_files = glob.glob("*.nc")  # order is 
```

```python
beg_yr_ind = ecco_files[0].find('mon_mean_') + 9
end_yr_ind = beg_yr_ind + 4

beg_mo_ind = end_yr_ind + 1
end_mo_ind = beg_mo_ind + 2

ecco_yrs = []
ecco_mos = []

for i in np.arange(0, len(ecco_files)):
    
    ecco_yrs.append(ecco_files[i][beg_yr_ind:end_yr_ind])
    ecco_mos.append(ecco_files[i][beg_mo_ind:end_mo_ind])

ecco_yrs = list(set(ecco_yrs))
ecco_mos = list(set(ecco_mos))

ecco_yrs.sort()
ecco_mos.sort()
```

```python
def ecco_name(yr, mo):
    return ecco_files[0][:beg_yr_ind ] + yr + '-' + mo + ecco_files[0][end_mo_ind:]
```

```python
ex_ecco = xr.open_dataset(ecco_dir + ecco_files[0])

ex_ecco.Z.data;
```

```python
# define some constants
ec_lon = ex_ecco.longitude.data
ec_lat = ex_ecco.latitude.data

Nx = ec_lon.shape[0]
Ny = ec_lat.shape[0]
Nz = ex_ecco.Z.data.shape[0]
```

```python
# compute climatological mean of depth-averaged velocity for first two years
u_mean = np.zeros((1,Ny,Nx))
v_mean = np.zeros((1,Ny,Nx))

count = 0.

with warnings.catch_warnings():
    warnings.simplefilter("ignore", category=RuntimeWarning)
    for yr in ecco_yrs[0:]:
        for mo in ecco_mos:
            ecco_ds = xr.open_dataset(ecco_dir + ecco_name(yr, mo))
            
            u_mean += np.nanmean(ecco_ds.EVEL.data[0,0:,:,:], axis=0)  # np.nanmean(ecco_ds.EVEL.data, axis=1)
            v_mean += np.nanmean(ecco_ds.NVEL.data[0,0:,:,:], axis=0) # np.nanmean(ecco_ds.NVEL.data, axis=1)
            
            count+=1

u_mean = u_mean[0,:,:]/count
v_mean = v_mean[0,:,:]/count
```

```python
fig, ax = plt.subplots(1,1, figsize=(24,10))

pc1 = ax.pcolormesh(ec_lon, ec_lat, u_mean, vmin=-0.05, vmax=0.05)

plt.colorbar(pc1)
```

# Interpolate bottom slope onto ECCO grid

```python
from scipy.interpolate import RectSphereBivariateSpline
```

```python
# pc1 = plt.pcolormesh(ec_lon_int)

# pc1 = plt.pcolormesh(np.flip(global_topo.detab_dx.data, axis=0))
# pc1 = plt.pcolormesh(global_topo.detab_dx.data, vmax=1e-3, vmin=-1e-3)
# plt.colorbar(pc1)

# plt.figure()
# plt.plot(lon.data * np.pi / 180)

```

```python
# interpolate dx of eta_b
ec_lat_int, ec_lon_int = np.meshgrid(-ec_lat* np.pi / 180 + np.pi/2,
                     ec_lon * np.pi / 180)

lut = RectSphereBivariateSpline(np.flip(-lat.data * np.pi / 180 + np.pi/2), lon.data * np.pi / 180, np.flip(global_topo.detab_dx.data, axis=0))

ec_detab_dx = np.transpose(lut.ev(ec_lat_int.ravel(), ec_lon_int.ravel()).reshape((len(ec_lon), len(ec_lat))))
```

```python
# interpolate dy of eta_b
ec_lat_int, ec_lon_int = np.meshgrid(-ec_lat* np.pi / 180 + np.pi/2,
                     ec_lon * np.pi / 180)

lut = RectSphereBivariateSpline(np.flip(-lat.data * np.pi / 180 + np.pi/2), lon.data * np.pi / 180, np.flip(global_topo.detab_dy.data, axis=0))

ec_detab_dy = np.transpose(lut.ev(ec_lat_int.ravel(), ec_lon_int.ravel()).reshape((len(ec_lon), len(ec_lat))))
```

```python
pre_interp_land_mask  = np.abs(global_topo.land_mask.data - 1)

lut = RectSphereBivariateSpline(np.flip(-lat.data * np.pi / 180 + np.pi/2), lon.data * np.pi / 180, np.flip(pre_interp_land_mask, axis=0))

ec_land_mask = np.round(np.transpose(lut.ev(ec_lat_int.ravel(), ec_lon_int.ravel()).reshape((len(ec_lon), len(ec_lat)))))

ec_land_mask[ec_land_mask==0] = np.nan
```

```python
ec_gradient = np.sqrt(ec_detab_dx**2 + ec_detab_dy**2) * ec_land_mask
```

```python
eps=1e-4

fig, ax = plt.subplots(1,1, figsize=(14,6))
pc1 = ax.pcolormesh(ec_lon, ec_lat, ec_gradient, norm=LogNorm(vmin=eps, vmax=0.05), shading='auto', cmap=plt.get_cmap('rainbow'))

plt.colorbar(pc1)
```
